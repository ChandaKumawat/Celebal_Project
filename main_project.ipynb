```python
X_train = df.drop(['Class','word_freq_table'], axis=1).values
Y_train = df['Class'].values
```

```python
print("Features shape (Independent) : ",X_train.shape)
print("Labels shape (Dependent) : ",Y_train.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(X_train, Y_train, test_size = 0.2, random_state = 5)
```

```python
print("X Train shape   :", x_train.shape)
print("X Test shape   :", x_test.shape)
print("Y Train shape   :", y_train.shape)
print("Y Test shape   :", y_test.shape)
```

```python
#Importing the library to implement Naive Bayesian Classification
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
#Train the model
model.fit(x_train,y_train)
```

```python
# Predicted labels for X train
y_pred_train= model.predict(x_train)
# Predicted labels for X test
y_pred_test= model.predict(x_test)
```

```python
#Accuracy score for train and test set in Train dataset
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)
print('Training Accuracy : ',train_accuracy)
print('Testing Accuracy : ',test_accuracy)
```

```python
#Implementing confusion matrix over it
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred_test)
print(cm)
```

```python
#Function to plot confusion matrix
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=30)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
#Plotting confusion matrix by calling above function
classes = ['Spam', 'Ham']
plot_confusion_matrix(cm, classes)
```

```python
TN = cm[0][0]
FN = cm[1][0]
FP = cm[0][1]
TP = cm[1][1]
Accuracy = (TP+TN)/(TP+FP+FN+TN)
Precision = (TP)/(TP+FP)
Recall = TP/(TP+FN)
f1 = (2*Precision*Recall)/(Precision+Recall)
print("Accuracy:", Accuracy)
print("Precision:", Precision)
print("Recall:", Recall)
print("F1 score:", f1)
```
